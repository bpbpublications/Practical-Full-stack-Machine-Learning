```python
!ls -wc data/train/ data/validation/ 
```

Downloaded dataset has 1000 labeled images for training and 800 labeled images for validation.

We would like to capture the dataset

```python
!dvc add data
```

<b>dvc add data</b> tells Git to ignore the directory and puts it into the DVC cache while keeping a file link to actual file. dvc file has the link to actual images. This command should be used on files or directories that are too large to be put into Git. Datasets and models are few examples of those large files

```python
#TIme to run the training
!python train.py
```

```python
#Time to capture the current version of the model. 
#This command produces a bunch of files, 
#among them model.h5 and metrics.json, weights of the trained model and metrics history
!dvc add model.h5
```

```python
#Commit the changes. 
#!git add .gitignore model.h5.dvc data.dvc metrics.json
#!git commit -m "First model, trained with 1000 images"
#!git tag -a "v1.0" -m "model v1.0, 1000 images"
```

### Understanding DVC

#### Files and directories

```python
!cat data.dvc
```

<b>path:</b> Path to the output, relative to the wdir path.

<b>md5</b>: md5 checksum for the output.

<b>cache</b>: Whether or not dvc should cache the output.

<b>metric</b>: Whether or not this file is a metric file.

```python
! ls -a
```

```python
! tree -L 1 .dvc
```

The cache directory will store the data. 
The data files and directories in DVC repositories will only contain links to the data files in the cache.
